```python
class Node:

    def __init__(self, value):
        self.data = value
        self.next = None
```

```python
class LinkedList:
    def __init__(self):
       # Create an Empty a linkedList
        self.head = None
        # Num. of nodes in the LinkedList
        self.n = 0

    def __len__(self):
        return self.n

    def insert_head(self,value):
        # new node
        new_node =Node(value)
        # create cunnection
        new_node.next = self.head
        # reassign head
        self.head = new_node
        #increment n
        self.n = self.n+1

    # def traverse(self):
    #     curr = self.head
    #     while curr != None:
    #         print(curr.data)
    #         curr = curr.next


    def __str__(self):
        curr = self.head
        result =""
        while curr != None:
            result = result + str(curr.data)+ "->"
            curr = curr.next
        return result[:-2]
```
