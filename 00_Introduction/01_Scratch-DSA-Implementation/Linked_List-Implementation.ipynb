# Linked List Implementation 

```python
class Node:

    def __init__(self,value):
        self.data = value
        self.next = None
```

```python
a = Node(1)
b = Node(2)
c = Node(3)
```

```python
print(a)
```

```python
print(a.data)
```

```python
print(a.next)
```

```python
a.next = b
b.next = c
```
